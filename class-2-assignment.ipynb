### 4. Calculate the Sum of Digits
**Task**: Given a number, find the sum of its digits.  
**Demo Input**: `123`  
**Demo Output**: `6`  
(Explanation: `1+2+3=6`)

```python
digit = input()

seperate_digits=list(digit)

sum=0

for i in seperate_digits:
    sum=sum+int(i)

print(sum)
```

### 5. Find Factorial of a Number
**Task**: Calculate the factorial of a given number.  
**Demo Input**: `4`  
**Demo Output**: `24`  
(Explanation: `4! = 4*3*2*1 = 24`)

```python
factorial=int(input())

multiply=1

for j in range(factorial,1,-1):
    multiply *=j

print(multiply)
```

### 6. Count Occurrences of a Digit
**Task**: Count how many times a specific digit appears max in a number.  
**Demo Input**:  1233321 <br>
**Demo Output**: `3`  <br>
(Explanation: `3` appears 3 times in `1233321`)

```python
def int_list(received_number): # this funtiun converts the list of strings to list of integers 
    int_number =  []
    for k in received_number:
        int_number.append(int(k))
    return int_number

def max_number(received_number): # this function counts how many times a specific digit appears max in a number
    max = 0
    curr_frequency = 0
    for m in received_number:
        curr_frequency = received_number.count(m)
        if(curr_frequency > max):
            max = curr_frequency
    return max

entered_number = list(input()) # input the number
int_number = int_list(entered_number) # called the function that converts string to int 
max = max_number(int_number) # called the function that counts how many times a specific digit appears max in a number 
print(max) 

#print(num)
#num=received_number[0]
#num = m
```

### 7. Find the GCD of Two Numbers
**Task**: Find the greatest common divisor (GCD) of two numbers.  
**Demo Input**: 8 and 12 <br>
**Demo Output**: `4`

```python
import math

def int_list(received_number): # this funtiun converts the list of strings to list of integers 
    int_number =  []
    for k in received_number:
        int_number.append(int(k))
    return int_number

gcd_numbers = input().split(' and ')
gcd=int_list(gcd_numbers)
print(math.gcd(gcd[0],gcd[1]))
```

### 9. Check Armstrong Number
**Task**: Check if a number is an Armstrong number (the sum of its digits raised to the power of the number of digits equals the number).  
**Demo Input**: `153`  
**Demo Output**: `153 is an Armstrong Number`  
(Explanation: \(1^3 + 5^3 + 3^3 = 153\))

```python
given_number = input()
input_number = int(given_number)
separate_numbers = list(given_number)
power = len(separate_numbers)
summation = 0

for p in separate_numbers:
    summation += pow(int(p),power)

if summation==input_number:
    print(input_number,'is an Armstrong Number')
else:
    print(input_number,'is not an Armstrong Number')
```

### 10. Generate a Pattern
**Task**: Print a pyramid pattern with `n` rows.  
**Demo input:** 4 <br>
**Output:** You will see the pyramid pattern with 4 rows

```python
rows = int(input())

for s in range(1,rows+1,1):
    print(" " * (rows - s),end="")
    print("*" * (2 * s - 1))

'''
Print (rows - s) spaces to align the stars in a pyramid shape.
Print (2 * s - 1) stars to create the pyramid layer for that row.
'''
```
